関数 f(x) = x^2 + 1 について、
制約条件 x >= -1 のもとで、f(x) を最小にする x を SciPy を使って求める。

```python
import numpy as np
from scipy.optimize import minimize

def objective(x):
    return x[0]**2 + 1

x0 = np.array([2.0])

bounds = [(0, None)]

answer_5_cm_3 = minimize(objective, x0, method='SLSQP', bounds=bounds)

print("answer_5_cm_3:")
print(answer_5_cm_3)

print("最小にする x:")
print(answer_5_cm_3.x)

print("最小値 f(x):")
print(answer_5_cm_3.fun)

print("制約条件 x >= -1 を満たしているか:")
print(answer_5_cm_3.x[0] >= -1)


```
