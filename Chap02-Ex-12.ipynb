0～1の値をとる一様乱数を1000個、2組発生させて、それぞれのヒストグラムを描いてみましょう。なお、それぞれのヒストグラムを別のグラフに表示させるために、plt.subplotを利用してください。ここで、一様乱数とは、ある数から、別のある数まで等確率で発生する乱数のことを言い、np.random.uniformを使います。例えば、0～1までの数を10個発生させる場合はnp.random.uniform(0.0,1.0,10)とします。
また、1000個だけじゃなくて、100個や10000個でも実施してみましょう。何かわかることはありますか？


```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.random.uniform(0.0, 1.0, 100)
data2 = np.random.uniform(0.0, 1.0, 10000)

plt.subplot(1, 2, 1)
plt.hist(data1, bins=20)
plt.title("data1")

plt.subplot(1, 2, 2)
plt.hist(data2, bins=20)
plt.title("data2")

plt.tight_layout()
plt.show()
```
