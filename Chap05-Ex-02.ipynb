0 <= x <= 1, 0 <= y <= 1 - x の三角領域で定義される関数
f(x, y) = 1 / ( sqrt(x + y) * (1 + x + y)^2 )
の積分値を求める。

```python
# 総合問題 5-2
import numpy as np
from scipy import integrate

answer_5_cm_2 = integrate.dblquad(
    lambda y, x: 1 / (np.sqrt(x + y) * (1 + x + y)**2),
    0,
    1,
    lambda x: 0,
    lambda x: 1 - x
)

print("answer_5_cm_2:")
print(answer_5_cm_2)

print("積分値:")
print(answer_5_cm_2[0])

print("推定誤差:")
print(answer_5_cm_2[1])
```
