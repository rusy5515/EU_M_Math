y=sin(x)とy=con(x)のグラフを重ねて書いてください。（ｘはー10～10）

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="y = sin(x)")
plt.plot(x, y2, label="y = cos(x)")

plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin(x) and y = cos(x)")
plt.grid(True)
plt.legend(loc="best")

plt.show()
```
