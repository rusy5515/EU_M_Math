## 1. k-means法

### 手法
k-means法は、あらかじめクラスタ数kを指定し、データをk個のグループに分けるクラスタリング手法である。
各クラスタの中心は、所属しているデータの平均値で表される。

### メリット
- 仕組みが単純で理解しやすい。
- 計算が比較的速い。
- Pythonのscikit-learnで簡単に実行できる。

### デメリット
- クラスタ数kを事前に決める必要がある。
- 初期値によって結果が変わることがある。
- 外れ値の影響を受けやすい。

### 実行方法

```python
from skleran.cluster import KMeans

model = KMeans(n_clusters=3)
model.fit(x)
labels = model.predict(x)
```

## 2. k-means++法

### 手法
k-means++法は、k-means法の初期中心の選び方を改良した方法である。
ランダムに中心を置くのではなく、できるだけ離れた位置に初期中心を配置することで、クラスタリング結果を安定させやすくする。

### メリット
- 通常のランダム初期化より結果が安定しやすい。
- 悪い初期値による失敗を減らせる。
- scikit-learnのKMeansで簡単に指定できる。

### デメリット
- クラスタ数kは自分で決める必要がある。
- 外れ値に強くなるわけではない。
- k-means法の弱点が完全になくなるわけではない。

### 実行方法

```python
from skleran.cluster import KMeans

model = KMeans(n_clusters=3, init='k-means++',random_state=42)
model.fit(x)
labels = model.predict(x)
```

## 3. k-medoids法

### 手法
k-medoids法は、クラスタの中心を平均値ではなく、実際に存在するデータ点から選ぶクラスタリング手法である。
この代表点をmedoidという。

### メリット
- 中心が実際のデータ点なので解釈しやすい。
- 平均値を使わないため、外れ値の影響を受けにくい。
- k-meansより代表点の意味が分かりやすい。

### デメリット
- k-means法より計算量が大きくなりやすい。
- クラスタ数kを事前に決める必要がある。
- scikit-learn本体ではなく、追加ライブラリが必要になる場合がある。

### 実行方法

```python
from skleran.cluster import KMedoids

model = KMedoids(n_clusters=3,random_state=0)
model.fit(x)
labels = model.predict(x)
```
