# 正則化項の目的とメリットについて

通常の重回帰では、実際の値 $y_i$ と予測値 $\hat{y}_i$ の誤差が小さくなるようにモデルを学習する。  
例えば、最小二乗法では次のような誤差の二乗和を小さくする。

$$
\sum_i (y_i - \hat{y}_i)^2
$$

しかし、説明変数が多い場合や、説明変数同士に強い相関がある場合、モデルが訓練データに過剰に適合してしまうことがある。  
このような状態を **過学習** という。

過学習が起きると、訓練データに対するスコアは高くなるが、未知のテストデータに対するスコアは低くなる可能性がある。  
つまり、学習に使ったデータにはよく当てはまるが、新しいデータに対する予測性能が低くなる。

そこで、回帰係数が大きくなりすぎないように制限を加える。  
この制限のために損失関数へ追加される項を **正則化項** という。

正則化項を加えることで、モデルの係数が極端に大きくなることを防ぎ、モデルを単純で安定したものにできる。  
その結果、訓練データだけに合わせすぎることを抑え、テストデータに対する予測性能の改善が期待できる。

---

## リッジ回帰

リッジ回帰では、係数の二乗和を正則化項として加える。

$$
\sum_i (y_i - \hat{y}_i)^2 + \alpha \sum_j w_j^2
$$

ここで、$w_j$ は回帰係数、$\alpha$ は正則化の強さを表すパラメータである。

リッジ回帰では、係数を全体的に小さくする効果がある。  
そのため、説明変数同士に強い相関がある場合でも、係数が極端に大きくなることを防ぎ、モデルを安定させやすい。

---

## ラッソ回帰

ラッソ回帰では、係数の絶対値の和を正則化項として加える。

$$
\sum_i (y_i - \hat{y}_i)^2 + \alpha \sum_j |w_j|
$$

ラッソ回帰では、一部の係数が $0$ になりやすいという特徴がある。  
そのため、重要でない説明変数を取り除く **変数選択** の効果がある。

説明変数が多い場合、すべての変数が予測に役立つとは限らない。  
ラッソ回帰を使うことで、予測にあまり寄与しない変数の係数を $0$ にし、より単純なモデルを作ることができる。

---

## 多重共線性について

多重共線性とは、説明変数同士に強い相関がある状態のことである。  
例えば、自動車データでは、車体が大きい車ほどエンジンサイズも大きい、というように、説明変数同士が似た情報を持っている場合がある。

多重共線性があると、通常の重回帰では係数が不安定になりやすい。  
つまり、少しデータが変わっただけで、各説明変数の係数が大きく変化する可能性がある。

正則化を用いると、係数の大きさに制限がかかるため、多重共線性による係数の不安定さを軽減できる。  
特にリッジ回帰は、説明変数同士に相関がある場合でも、係数を全体的に小さくしてモデルを安定させやすい。

---

## 過学習について

過学習とは、モデルが訓練データに合わせすぎてしまい、未知のデータに対する予測性能が低下することである。

例えば、訓練データのスコアが高いのに、テストデータのスコアが低い場合、過学習が起きている可能性がある。  
これは、モデルが訓練データの細かい特徴やノイズまで学習してしまい、一般的な傾向をうまく捉えられていない状態である。

正則化を使うと、係数が大きくなりすぎることを防げるため、モデルが複雑になりすぎることを抑えられる。  
そのため、過学習を防ぐ効果がある。

---

## パラメータ選択について

リッジ回帰やラッソ回帰では、正則化の強さを表すパラメータとして `alpha` を指定する。

`alpha` が小さい場合、正則化の効果は弱くなり、通常の重回帰に近いモデルになる。  
一方で、`alpha` が大きい場合、係数への制限が強くなり、モデルはより単純になる。

ただし、`alpha` を大きくしすぎると、必要な説明変数の影響まで小さくしてしまい、予測性能が下がる可能性がある。  
逆に、`alpha` が小さすぎると、正則化の効果が弱く、過学習を十分に防げない可能性がある。

したがって、正則化を使う場合は、訓練データのスコアだけでなく、テストデータのスコアも確認しながら、適切な `alpha` を選ぶ必要がある。

---

## まとめ

正則化項の目的は、回帰係数が大きくなりすぎることを防ぎ、モデルを単純で安定したものにすることである。  
正則化には、過学習を抑える、係数を安定させる、多重共線性の影響を軽減する、というメリットがある。

リッジ回帰では係数の二乗和を用いて、係数を全体的に小さくする。  
ラッソ回帰では係数の絶対値の和を用いて、一部の係数を $0$ にしやすくする。

今回の課題では、自動車価格データを用いて、`width` と `engine-size` から `price` を予測した。  
重回帰では説明変数と目的変数の線形関係を学習し、ラッソ回帰では正則化項を加えることで、係数が大きくなりすぎることを防いだ。

正則化を使うときは、`alpha` の値を適切に選ぶことが重要である。  
`alpha` が大きすぎるとモデルが単純になりすぎ、逆に小さすぎると過学習を十分に防げない。  
そのため、訓練データとテストデータのスコアを比較しながら、モデルの汎化性能を確認することが重要である。
