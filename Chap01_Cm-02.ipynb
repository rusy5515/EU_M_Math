### 設問
上記をさらに一般化して、Nの自然数として、Nまでの素数を表示する関数を書いてください。

```python


def print_primes(N):

    for n in range(2, N+1):
        is_prime = True

        for i in range(2, n):
            if n % i == 0:
                is_prime = False
        
        if is_prime == True:
            print(n)    
```

```python
print_primes(30)
```
