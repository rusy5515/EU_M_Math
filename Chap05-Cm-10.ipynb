5-3 の補間のサンプルコードを参考にして、最近傍補間を実行し、グラフを描きなさい。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.linspace(0, 10, 11, endpoint=True)
y = np.sin(x**2/5.0)
plt.plot(x, y, 'o' )
plt.grid(True)
plt.show()

xnew = np.linspace(0, 10, num=101, endpoint=True)

f = interpolate.interp1d(x, y, kind='nearest')

plt.plot(x, y, 'o', xnew, f(xnew), '-')
plt.grid(True)
plt.show()
```
