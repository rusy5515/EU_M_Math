# 総合問題 8-1

## 回帰
回帰とは、連続値を予測する教師あり学習の方法である。例えば、住宅価格、売上、気温、株価などの数値を予測する場面で使われる。

## 分類
分類とは、データをあらかじめ決められたカテゴリに分ける教師あり学習の方法である。例えば、メールが迷惑メールかどうか、腫瘍が悪性か良性か、キノコが毒か食用かを判定する場面で使われる。

## 教師あり学習
教師あり学習とは、入力データと正解ラベルの組を使ってモデルを学習する方法である。説明変数Xから目的変数yを予測できるように学習する。回帰と分類は教師あり学習の代表例である。

## 重回帰分析
重回帰分析とは、複数の説明変数を用いて連続値を予測する回帰分析である。例えば、部屋の広さ、築年数、駅からの距離などを使って家賃を予測する場合に使える。

## ロジスティック回帰分析
ロジスティック回帰分析とは、主に分類問題に使われるモデルである。説明変数に係数をかけて足し合わせたスコアを作り、それを確率に変換して0か1かを分類する。例えば、年収が50Kを超えるか、病気であるかどうかを予測する場合に使える。

## 正則化
正則化とは、モデルの係数が大きくなりすぎないように制限をかける方法である。モデルが訓練データに過剰に適合する過学習を防ぐ目的で使われる。

## リッジ回帰
リッジ回帰とは、回帰モデルにL2正則化を加えた手法である。係数の二乗和にペナルティをかけることで、係数が極端に大きくなることを防ぐ。

## ラッソ回帰
ラッソ回帰とは、回帰モデルにL1正則化を加えた手法である。不要な説明変数の係数を0にしやすいため、特徴量選択の効果も持つ。

## 決定木
決定木とは、条件分岐を繰り返して分類や回帰を行うモデルである。例えば「においが強いか」「色が特定の色か」などの条件を使って、データを枝分かれさせながら予測する。解釈しやすい一方で、木を深くしすぎると過学習しやすい。

## エントロピー
エントロピーとは、分類におけるクラスの混ざり具合を表す指標である。1つのグループに複数のクラスが半々で混ざっている場合、エントロピーは高い。逆に、1つのクラスだけに偏っている場合、エントロピーは低い。

## 情報利得
情報利得とは、ある条件でデータを分割したときに、エントロピーがどれだけ下がったかを表す指標である。情報利得が大きい分割ほど、分類に役立つ良い条件分岐である。

## k-NN法
k-NN法とは、k近傍法とも呼ばれ、新しいデータに対して近くにあるk個のデータを見て分類や回帰を行う方法である。近くのデータの多数決によって分類するため、直感的で分かりやすいが、データ量が多いと計算コストが大きくなる。

## SVM
SVMとは、サポートベクターマシンのことで、分類境界とデータ点との距離、つまりマージンをできるだけ大きくするように境界を決める分類モデルである。特徴量のスケールの影響を受けやすいため、標準化が重要である。

## ノーフリーランチ
ノーフリーランチとは、すべての問題に対して常に最良となる万能な機械学習モデルは存在しない、という考え方である。ある問題で高性能なモデルが、別の問題でも必ず高性能とは限らない。そのため、データや目的に応じて複数のモデルを試し、比較する必要がある。
