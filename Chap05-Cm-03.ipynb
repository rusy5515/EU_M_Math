次の x_array, y_array があるとき、NumPy の where を用いて、
3番目と4番目は x_array から、
1番目、2番目、5番目は y_array から取り出したデータを生成しなさい。

```python
import numpy as np

x_array = np.array([1,2,3,4,5])
y_array = np.array([6,7,8,9,10])

cond_data = np.array([False,False,True,True,False])


answer_5_3 = np.where(cond_data, x_array, y_array)
print("answer_5_3")
print(answer_5_3)
```
