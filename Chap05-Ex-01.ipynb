次の行列 A とベクトル b に対して、コレスキー分解を用いて Ax = b の方程式を解く。

```python
import numpy as np
from scipy import linalg

np.set_printoptions(precision=3, suppress=True)

A = np.array([
    [5, 1, 0, 1],
    [1, 9, -5, 7],
    [0, -5, 8, -3],
    [1, 7, -3, 10]
], dtype=float)

b = np.array([2,10,5,10], dtype=float)

print("A:")
print(A)

print("b:")
print(b)

R = linalg.cholesky(A)

print("R:")
print(R)


y = linalg.solve(R.T, b)
x = linalg.solve(R, y)

print("x:")
print(x)

print("A @ x:")
print(A @ x)

print("A @ x と b がほぼ一致するか:")
print(np.allclose(A @ x, b))
```
