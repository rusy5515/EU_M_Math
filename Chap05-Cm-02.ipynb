sample_names が 'c' ではない data を取り出しなさい。

```python
import numpy as np
import numpy.random as random

sample_name = np.array(['a', 'b', 'c', 'd', 'a'])

random.seed(0)
data = random.randn(5,5)

print("sample_name")
print(sample_name)

print("data")
print(data)

answer_5_2 = data[sample_name != 'c']
print("answer_5_2")
print(answer_5_2)

```
