```python
from pandas import Series, DataFrame
import pandas as pd

attri_data1 = {
    'ID': ['1', '2', '3', '4', '5'],
    'sex': ['F', 'F', 'M', 'M', 'F'],
    'Money': ['1000', '2000', '500', '300', '700'],
    'Name': ['Saito', 'Horie', 'kondo', 'Kawada', 'Matubara']
}

attri_data2 = {
    'ID': ['3', '4', '7'],
    'Math': [60, 30, 40],
    'English': [80, 20, 30],
}

attri_data_frame1 = DataFrame(attri_data1)
attri_data_frame2 = DataFrame(attri_data2)

merge_data = pd.merge(attri_data_frame1, attri_data_frame2, on='ID')

merge_data['Money'] = merge_data['Money'].astype(int)

result = merge_data[['Money', 'Math', 'English']].mean()

print('Merged data:')
print(merge_data)

print('\nAverage:')
print(result)
```
