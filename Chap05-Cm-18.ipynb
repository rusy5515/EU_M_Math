関数 f(x) = x^3 - 2x^2 - 11x + 12 について、
f(x) = 0 となる x の値を fsolve を用いて求める。

```python
import numpy as np

coefficients = [1, -2, -11, 12]

answer_5_17 = np.roots(coefficients)
print("answer_5_17", answer_5_17[0])
print("f(answer_5_17)", np.polyval(coefficients, answer_5_17[0]))
```

```python
# 練習問題 5-18
from scipy.optimize import fsolve
import numpy as np

def f(x):
    return x**3 - 2*x**2 - 11*x + 12

answer_5_18_1 = fsolve(f, -5)
answer_5_18_2 = fsolve(f, 0)
answer_5_18_3 = fsolve(f, 5)

print("初期値 -5 から求めた解:")
print(answer_5_18_1)
print("f(answer_5_18_1):")
print(f(answer_5_18_1))

print("初期値 0 から求めた解:")
print(answer_5_18_2)
print("f(answer_5_18_2):")
print(f(answer_5_18_2))

print("初期値 5 から求めた解:")
print(answer_5_18_3)
print("f(answer_5_18_3):")
print(f(answer_5_18_3))
```
