Chap02-Ex07のデータに対して男女別（MF別）の平均moneyを求めよ。

```python
from pandas import Series, DataFrame
import pandas as pd

attri_data1 = {
    'ID': ['1', '2', '3', '4', '5'],
    'sex': ['F', 'F', 'M', 'M', 'F'],
    'Money': ['1000', '2000', '500', '300', '700'],
    'Name': ['Saito', 'Horie', 'kondo', 'Kawada', 'Matubara']
}

attri_data_frame = DataFrame(attri_data1)

attri_data_frame['Money'] = attri_data_frame['Money'].astype(int)

result = attri_data_frame.groupby('sex')['Money'].mean()

print(result)
```
