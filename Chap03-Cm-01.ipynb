【総合問題3-1 統計の基礎と可視化】
以下のサイトにあるデータ（ワインの品質）を読み込み、以下の問いに答えてください。

http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

1,要約統計量（平均、最大値、最小値、標準偏差など）を算出してください。なお、pandasには、データをアウトプットできるメソッド（to_csv）もありますので、余裕があれば、計算した基本統計量の結果をCSVファイルに保存するところまでやってみましょう。

2,それぞれの変数の分布と、それぞれの変数の関係性（2変数間のみ）がわかるように、グラフ化してみましょう。すべての変数を用いて実行すると時間がかかりますので、注意しましょう。何かわかる傾向はありますか。

```python
import pandas as pd
import matplotlib.pyplot as plt

wine = pd.read_csv("winequality-red.csv",sep=";")

wine.shape
wine.columns

summary = wine.describe()

summary

summary.to_csv("summary.csv")

wine.hist(figsize=(14,10),bins=20)
plt.tight_layout()
plt.show()

cols = ["alcohol", "volatile acidity", "citric acid", "sulphates", "quality"]

wine[cols].hist(figsize=(12, 8), bins=20)
plt.tight_layout()
plt.show()

plt.scatter(wine["alcohol"], wine["quality"])
plt.xlabel("alcohol")
plt.ylabel("quality")
plt.title("alcohol and quality")
plt.grid(True)
plt.show()


```

要約統計量を確認すると、quality は 3 から 8 の範囲で、平均は約 5.64 であった。
ヒストグラムを見ると、quality は 5 と 6 に集中している。
相関係数とグラフから、alcohol は quality と比較的強い正の相関があり、
volatile acidity は quality と負の相関があることが分かった。
したがって、アルコール度数が高いワインほど品質評価が高く、
揮発酸が高いワインほど品質評価が低くなる傾向があると考えられる。
