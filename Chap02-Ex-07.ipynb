以下のデータに足して、moneyが500以上の人を絞り込んで、レコードに表示してください。

```python
import pandas as pd
from pandas import Series,DataFrame

attri_data1 = {'ID':['1','2','3','4','5'],
                'sex':['F','F','M','M','F',],
                'Money':['1000','2000','500','300','700'],
                'Name':['Saito','Horie','kondo','Kawada','Matubara']}
attri_data_frame = DataFrame(attri_data1)

attri_data_frame['Money'] = attri_data_frame['Money'].astype(int)

result = attri_data_frame[attri_data_frame['Money'] >= 500]

print(result)
```
